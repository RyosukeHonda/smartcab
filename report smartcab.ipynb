```python
import  pandas as pd
df100=pd.read_csv('report.csv')
def success_rate(df):
    return (100.0-len(df[df["Deadline"]==0]))/100.0
def total_reward(df):
    time0=df[df['Time']==0]["Unnamed: 0"]
    df_time0=df.ix[time0-1]
    df_time0=df_time0.drop(df_time0.index[0])
    total_reward=pd.merge(df_time0,df.tail(1),how='outer')
    plt.plot(total_reward['Episode'],total_reward['Reward'])
    plt.xlabel('No.Episode')
    plt.ylabel('Total Reward')
   
    return total_reward['Reward'].cumsum().tail(1)
```

```python
df_constant=pd.read_csv('constant.csv')
```

```python
success_rate(df_constant)
```

```python
total_reward(df_constant)
```

```python
total_reward(df100)
```

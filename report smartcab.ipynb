```python
import  pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df100=pd.read_csv('report.csv')
def success_rate(df):
    return (100.0-len(df[df["Deadline"]==0]))/100.0
def total_reward(df):
    time0=df[df['Time']==0]["Unnamed: 0"]
    df_time0=df.ix[time0-1]
    df_time0=df_time0.drop(df_time0.index[0])
    total_reward=pd.merge(df_time0,df.tail(1),how='outer')
    plt.plot(total_reward['Episode'],total_reward['Total_Reward'])
    plt.xlabel('No.Episode')
    plt.ylabel('Total Reward')
    return total_reward['Total_Reward'].cumsum().tail(1)

def negative_action_rate(df):
    return 1.0*len(df[df['Reward']<0])/len(df)
  
   
    
```
